# Clase 2 - Reconocimiento de Patrones en Imagenes
En este notebook, se realiza un análisis básico de imágenes utilizando la biblioteca OpenCV. Se selecciona una región de interés (ROI) de una imagen, se realizan operaciones de umbralización y se exploran diferentes representaciones de color. Este proceso es útil para tareas como segmentación de imágenes y análisis de características.

```python
# Importar las bibliotecas necesarias para el procesamiento de imágenes y visualización
import matplotlib.pyplot as plt
import cv2
import numpy as np

# Leer la imagen desde el archivo especificado
img = cv2.imread('../IMG/lena.png')

#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Seleccionar una región de interés (ROI) de la imagen
# En este caso, se selecciona un área específica del canal azul
roi = img[255:282, 250:280, 0]
```

```python
# Visualizar la región de interés (ROI) seleccionada
plt.figure()
plt.imshow(roi , cmap = 'gray')
plt.show()

# Imprimir los valores de los píxeles de la ROI para inspección
print(roi)
```

```python
# Aplicar un umbral a la ROI para identificar píxeles con valores mayores a 200
print(roi>200)

# Crear una máscara binaria donde los píxeles mayores a 150 se establecen en 1
temp = ((roi > 150)*1).astype(np.uint8)

# Visualizar la máscara binaria generada
plt.figure()
plt.imshow(temp , cmap = 'gray')
plt.show()
```

```python
# Convertir la imagen al espacio de color HSV (Tono, Saturación, Valor)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Dividir la imagen HSV en sus tres canales: H, S y V
h,s,v = cv2.split(hsv)

# Visualizar el canal de Tono (H)
plt.figure()
plt.imshow(h,cmap='gray')

# Visualizar el canal de Saturación (S)
plt.figure()
plt.imshow(s,cmap='gray')

# Visualizar el canal de Valor (V)
plt.figure()
plt.imshow(v,cmap='gray')

# Mostrar todas las figuras
plt.show()
```
