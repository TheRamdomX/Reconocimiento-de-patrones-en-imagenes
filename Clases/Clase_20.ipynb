## Metodo newton-raphson

Se quiere encontrar la aproximacion de una raiz

X = $ \sqrt{m} $

x² = m

x² - m = 0

```python
def f(x):
    return x**2 - m

def f_prime(x):
    return 2*x

def newton_raphson(x0, max_iter=100):
    x = x0
    for i in range(max_iter):
        print(f"raiz es {x}")
        x_new = x - f(x) / f_prime(x)
        if abs(x_new - x) < 1e-7:
            return x_new
        x = x_new
    return x

# valor cuyo raiz cuadrada queremos encontrar
m = 9

# valor inicial (arbitrario)
x0 = 4.5

raiz = newton_raphson(x0)
print(f"La raiz cuadrada de {m} es aproximadamente {raiz}")

print("\nProbando con otro valor inicial:\n")

raiz = newton_raphson(10)
print(f"La raiz cuadrada de {m} es aproximadamente {raiz}")
```

## BackPropagation

```python
import numpy as np
import matplotlib.pyplot as plt

e = np.exp(1)

# peso inicial
w = 0.69

# valor de entrada
x1 = 0.9

# valor objetivo
g = 0.5

# tasa de aprendizaje
lr = 0.2


def backpropagation(w_old, x1, g, lr):

    # salida de la neurona
    o = 1 / (1 + e**(-w_old * x1))

    # actualizacion del peso
    W_new = w_old + lr * (g - o) * o * (1 - o) * x1


    costo = 0.5 * (g - o)**2

    return W_new , costo

c_hist = []

print(f"Peso inicial: {w}, salida inicial: {1 / (1 + e**(-w * x1))}")
for i in range(1000):
    w, c = backpropagation(w, x1, g, lr)
    o = 1 / (1 + e**(-w * x1))
    print(f"Iteracion {i+1}: peso = {w}, salida = {o}, costo = {c}")
    c_hist.append(c)

plt.plot(c_hist)
plt.xlabel("Iteración")
plt.ylabel("Costo (Error cuadrático)")
plt.title("Convergencia del entrenamiento")
plt.show()

```
