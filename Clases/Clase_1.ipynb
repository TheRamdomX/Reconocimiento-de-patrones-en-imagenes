# Clase 1 - Reconocimiento de Patrones en Imagenes

1. **Carga de la Imagen**:
   - Se utiliza OpenCV para cargar una imagen llamada `lena.png` desde la carpeta `IMG`.
   - La imagen se carga en formato BGR (azul, verde, rojo).

2. **Conversión de Formato de Color**:
   - La imagen se convierte de BGR a RGB para que los colores se muestren correctamente al usar Matplotlib.

3. **Cálculo de Histogramas**:
   - Se calculan los histogramas de los tres canales de color (rojo, verde y azul) de la imagen.
   - Los histogramas muestran la distribución de intensidades de los píxeles para cada canal.

4. **Visualización de Histogramas**:
   - Los histogramas de los tres canales se grafican en una sola figura, con colores correspondientes a cada canal.

5. **Acceso a un Píxel Específico**:
   - Se accede al valor del píxel en la posición `(265, 327)` de la imagen.
   - Este valor contiene las intensidades de los canales rojo, verde y azul en esa posición.


```python
import cv2 
import matplotlib.pyplot as plt

# Load an image using OpenCV
img = cv2.imread('../IMG/lena.png')

# Convert BGR to RGB 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 

#red, green, blue = cv2.split(img)

hist_red = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist_red, color='red')

hist_green = cv2.calcHist([img],[1],None,[256],[0,256])
plt.plot(hist_green, color='green')

hist_blue = cv2.calcHist([img],[2],None,[256],[0,256])
plt.plot(hist_blue, color='blue')

plt.show()  

pixel = img[265,327,:]
print(pixel)
```
